```python
# 1️. Write a program using function to find greatest of three numbers.

# 2. Write a Python program using function to convert Celsius to Fahrenheit.

# 3. Write a Python function to remove a given word from a list and strip it at the same time.
```

```python
#1
def greatest_of_three(a, b, c):
    if a >= b and a >= c:
        return a
    elif b >= a and b >= c:
        return b
    else:
        return c

num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))
num3 = int(input("Enter third number: "))
greatest = greatest_of_three(num1, num2, num3)

print(f"The greatest number is {greatest}")
```

```python
#2
def celsius_to_fahrenheit(celsius):
    return (celsius * 9/5) + 32


c = float(input("Enter temperature in Celsius: "))

f = celsius_to_fahrenheit(c)

print(f"{c}°C is equal to {f}°F")
```

```python
#3
def remove_word_and_strip(word_list, word_to_remove):

    return [w.strip() for w in word_list if w.strip() != word_to_remove]

words = ["  apple", "banana  ", "  cherry ", "banana", " mango "]

word_to_remove = input("Enter the word to remove: ")

new_list = remove_word_and_strip(words, word_to_remove)

print("Updated list:", new_list)

```
